```python
import requests
```

```python
html_text = requests.get('https://www.cisa.gov/uscert/ncas/bulletins/sb22-234').text
soup = BeautifulSoup(html_text,'lxml')
print(soup)
```

```python
all_tables = soup.find('div',{'class':'main-container container js-quickedit-main-content'})
all_tables
```

```python
table = all_tables.find('table',{'id':'table_severity_not_yet_assigned'})
table
```

```python
description = []
for i in table.find_all('tr')[1:]:
  j = i.find_all('td')
  description.append(j[1].text)
```

```python
cve = []
for i in table.find_all('tr')[1:]:
  j = i.find_all('td')[-1]
  k = j.find_all('a')[0]
  cve.append(k.text)
```

```python
len(description)
```

```python
len(cve)
```

```python
import itertools
for (i,j) in itertools.zip_longest(cve, description):
    print (i,j)
```
