```python
#Creating a subsample of the dataset, becasue the dataset is too big to process
dataset_subset = dataset.sample(n = 50000, replace = False)
```

```python
subset = dataset_subset.head(50)
subset
```

```python
#splitting the word in ngrams
def splitting_in_ngrams(input_word):
    input_word=str(input_word)
    s=''
    for i in range(len(input_word)-2):
        s1 = input_word[i:i+3]
        s = s + ' ' + s1 
    s = s[1:]
    return(s)
```

```python
#creating a corpus of documents. A document is the list of ngrams extracted from a domain
corpus =[]
for i in range(len(dataset_subset)):
    temp = dataset_subset.iloc[i][1]
    temp = str(temp)
    temp = temp.lower()
    temp = splitting_in_ngrams(temp)
    corpus.append(temp)
print(corpus)
```

```python
# Creating the Bag of Words model on a smaller vocabulary to understand how it works
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 3)
cv.fit_transform(['bam amw mwa wap apl plo lof off ffx lon', 'lon ong ngw gwi win']).toarray()

for item in cv.fit_transform(['bam amw mwa wap apl plo lof off ffx lon', 'lon ong ngw gwi win']).toarray():
    print(item)
    
print(cv.transform(['bam amw']).toarray())
```

```python
X_array = subset.iloc[:,1].values
X_array
```

```python
#Splitting the dataset into grams
X_gram =[]
for i in range(len(subset)):
    temp = subset.iloc[i][1]
    temp = str(temp)
    temp = temp.lower()
    temp = splitting_in_ngrams(temp)
    X_gram.append(temp)
```

```python
#Transforming the dataset into a CountVector
X_vector = cv.fit_transform(X_gram).toarray()
X_vector
```

```python
#Label encoding y
y = subset.iloc[:,0].values
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_vector, y, test_size = 0.20, random_state = 0)
```

```python
# Training the Logistic Regression model on the Training set
from sklearn.linear_model import LogisticRegression
LG_classifier = LogisticRegression(random_state = 0)
classifier = LG_classifier.fit(X_train, y_train)
```

```python
#saving the Logistic Regression model parameters
import json
classification_model_param = {}
classification_model_param['coef'] = list(classifier.coef_)
classification_model_param['intercept'] = list(classifier.intercept_)

coeff = list(classifier.coef_)
intercept = list(classifier.intercept_)
print(coeff)
print(intercept)
```

```python
#predicting y_test using coeff and intercept. Trying to understand how these parameters are used in the LG_classifier
import numpy as np
from sklearn.utils.extmath import safe_sparse_dot

print(coeff)
print(intercept)
print(X_test)

out = np.matmul(X_test,coeff[0])

out_1 = safe_sparse_dot(X_test,coeff[0]) + intercept
print(out_1)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
print(y_pred)
print(y_test)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
