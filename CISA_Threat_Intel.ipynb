```python
import requests
```

```python
html_text = requests.get('https://www.cisa.gov/uscert/ncas/alerts/aa22-181a').text
soup = BeautifulSoup(html_text,'lxml')
print(soup)
```

```python
IOC = soup.find('div', {'class':'field field--name-field-aa-technical-details field--type-text-long field--label-hidden field--item'})
IOC
```

```python
all_tables = IOC.find_all('table', class_ = 'general-table')
```

```python
all_tables
```

```python
for i in all_tables[:-1]:
  print(i.text)
```
